```python
'''
Come up with a formula that returns how much of the task is left after day n, where n=1 is Jan 1.

After day 0 (so before we begin working on the task), the full task is pending, i.e. 100% = 1 is left.
f_0 = 1

On day n, we do 2 / (365 - n + 1) of what is left, which is f_(n-1), so at the end of day n, the remainder
of the task left to do is:

New remainder = How much was left before - how much we did today
f_n           = f_(n-1)                  - 2 / (365 - n + 1) * f_(n-1)

A bit of algebra:

f_n = f_(n-1) * (364 - n) / (366 - n)
'''

def left_after_day_n(n):
    if n == 0:
        return 1
    tmp = left_after_day_n(n-1)
    return tmp * (364 - n) / (366 - n)

n = 0
left_to_do = left_after_day_n(n)
while left_to_do > 0.00000000001:
    print(f"After day {n} there is {left_to_do} left to do")
    n += 1
    left_to_do = left_after_day_n(n)

print(f"After day {n} there is {left_to_do} left to do")
```
